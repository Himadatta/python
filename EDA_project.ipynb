Exploratory Data Analysis (EDA) is a fundamental step in data analysis that involves summarizing and visualizing datasets to uncover underlying patterns, relationships, and anomalies. EDA assists in revealing the distribution and any trends of the data by utilizing descriptive statistics and a variety of visualization techniques, such as histograms, scatter plots, and box plots. This initial analysis gives a clear knowledge of the dataset, making it possible to discover outliers, clean up the data, and provide some initial hypotheses.

EDA provides valuable insights by pointing out pertinent features, evaluating relationships, and recommending suitable modeling strategies for additional research. By emphasizing important data structures and properties, it also aids in the selection of the appropriate models and methodologies. All things considered, EDA is essential for making wise choices and laying the groundwork for more complex data analysis and machine learning assignments.

# Analyzing and Visualizing Experimental Data

It would be beneficial to comprehend the data distribution, including the countries where orders were shipped, the fulfillment kinds, the sales channels, and the product categories sold, before we start asking questions about the sales data. Investigating these factors is crucial to determining the dataset's representativeness of the larger Amazon sales environment. This size of dataset typically has some selection bias.

First, let's import seaborn and matplotlib.pyplot.

# **Monthly Sales Distribution`**

```python
monthly_sales = amazon_data.groupby('month')['orderAmount (INR)'].sum()
plt.figure(figsize=(10, 5))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values,hue=monthly_sales.index, palette='magma')
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

# **Monthly Sales Distribution by Category**

```python
plt.figure(figsize=(20, 5))
sns.barplot(x=amazon_data.month, y=amazon_data['orderAmount (INR)'], hue=amazon_data.productCategory, palette='coolwarm')
plt.title('Monthly Sales Distribution by Category')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='productCategory')
plt.show()
```

# **Order Status Distribution**

```python
order_status_dist =amazon_data['shipStatus'].value_counts()
plt.figure(figsize=(8, 6))
order_status_dist.plot.pie(autopct='%1.0f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Order Status Distribution')
plt.show()
```

# **Fulfillment Method Analysis**

```python
fulfillment_method = amazon_data.groupby('Fulfilment')['orderQuantity'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=fulfillment_method.index, y=fulfillment_method.values, palette='pastel')
plt.title('Fulfillment Method Analysis')
plt.xlabel('Fulfillment Method')
plt.ylabel('Total Quantity')
plt.show()
```

# **Order Cancelled**

```python
cancelled_orders = amazon_data[amazon_data['shipStatus'] == 'Cancelled']
cancelled_reasons = cancelled_orders.groupby('productCategory')['orderQuantity'].count().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
cancelled_reasons.plot.bar(color='lightgreen')
plt.title('Order Cancelled')
plt.xlabel('productCategory')
plt.ylabel('Number of Cancelled Orders')
plt.show()
```

# **customerType vs. Customer Sales**

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=amazon_data.customerType, y=amazon_data['orderAmount (INR)'], palette='Set2')
plt.title('customerType vs. Customer Sales')
plt.xlabel('Customer Type')
plt.ylabel('Amount')
plt.show()
```

# **Average Amount per Order by Fulfillment Method**

```python
avg_amount_order = amazon_data.groupby('Fulfilment')['orderAmount (INR)'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_amount_order.index, y=avg_amount_order.values, palette='Blues_r')
plt.title('Average Amount per Order by Fulfillment Method')
plt.xlabel('Fulfillment Method')
plt.ylabel('Average Amount')
plt.show()
```

# **Pairwise Relationships in the Dataset**

```python
sns.pairplot(amazon_data[['orderQuantity', 'orderAmount (INR)', 'date']], palette='husl')
plt.suptitle('Pairwise Relationships in the Dataset', y=1.02)
plt.show()
```

# **Customer Retention Over Months**

```python
customer_retention = amazon_data[amazon_data['shipStatus'] == 'Shipped - Delivered to Buyer'].groupby('month')['orderid'].nunique()
plt.figure(figsize=(10, 6))
sns.lineplot(x=customer_retention.index, y=customer_retention.values, marker='o', color='green')
plt.title('Customer Retention Over Months')
plt.xlabel('Month')
plt.ylabel('Number of Unique Customers')
plt.show()
```

# **Category-wise Average Transaction Amount**

```python
avg_transaction_amount = amazon_data.groupby('productCategory')['orderAmount (INR)'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=avg_transaction_amount.index, y=avg_transaction_amount.values, palette='coolwarm')
plt.title('Category-wise Average Transaction Amount')
plt.xlabel('Category')
plt.ylabel('Average Transaction Amount')
plt.xticks(rotation=45)
plt.show()
```
